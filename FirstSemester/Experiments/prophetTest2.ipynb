```python
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
dataFile = pd.read_csv('./GM.csv')
dataFile.head()
```

```python
dataFile = dataFile.drop(columns=["Open", "High", "Low", "Adj Close", "Volume"])
dataFile = dataFile.rename(columns={"Date": "ds", "Close": "y"})
dataFile.head()
```

```python
m = Prophet()
m.fit(dataFile)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
plot_plotly(m, forecast)
```

```python
plot_components_plotly(m, forecast)
```
